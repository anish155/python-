```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.thebritishcollege.edu.np/programme"
```

```python
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.text,'html')
print(soup)
```

```python
courses=soup.find_all("div",attrs={"class":"course-title d-flex align-items-center"})
courses
```

```python
module=[]
for x in courses:
    module.append( x.text.strip())
```
